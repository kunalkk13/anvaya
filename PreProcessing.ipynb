### Convert a .csv file into separate .txt files (Column-wise)

```python
import pandas as pd

read_file = 'train_anvaya_iast.csv'
write_file1 = 'train-anvaya-iast-shloka.txt'
write_file2 = 'train-anvaya-iast-prose.txt'
df = pd.read_csv(read_file)

# Ensuring the csv files have 2 columns
if df.shape[1]!=2:
    raise ValueError('The csv file should have exactly 2 columns')

# Extract the columns
col1, col2 = df.iloc[:, 0], df.iloc[:, 1]

col1.to_csv(write_file1, index=False, header=False)
col2.to_csv(write_file2, index=False, header=False)
```

### Merge two .txt files to a .csv file

```python
import pandas as pd

# Define the file paths
file_shloka = "train-anvaya-shloka-byt5Segment.txt"  
file_prose = "train-anvaya-prose-byt5Segment.txt"  
output_csv_file = "train_anvaya_byt5Segmented.csv"

# Read the contents of both text files
with open(file_shloka, "r") as f1:
    sloka_lines = f1.read().splitlines()

with open(file_prose, "r") as f2:
    prose_lines = f2.read().splitlines()

# Ensure both files have the same number of lines
if len(sloka_lines) != len(prose_lines):
    raise ValueError("Error: The two text files have a different number of lines!")

# Create a DataFrame
df = pd.DataFrame({"Sloka": sloka_lines, "Prose": prose_lines})

# Save to CSV
df.to_csv(output_csv_file, index=False)

print(f"✅ Successfully merged the text files into '{output_csv_file}'.")
```

### Substring match of csv files

```python
import pandas as pd
from fuzzywuzzy import fuzz


input_file = "test_anvaya.csv"  
output_file = "test-subs-match-pos.csv"
df = pd.read_csv(input_file)

# Assuming the columns are named 'Sloka' and 'Prose'
if 'Sloka' not in df.columns or 'Prose' not in df.columns:
    raise ValueError("Input CSV must have 'Sloka' and 'Prose' columns")


def find_partial_match_start(word1, word2):
    max_ratio = 0
    start_index = -1

    for i in range(len(word2) - len(word1) + 1):
        substring = word2[i:i+len(word1)]
        ratio = fuzz.ratio(word1, substring)
        if ratio > max_ratio:
            max_ratio = ratio
            start_index = i

    return start_index if max_ratio > 0 else None


def match_and_retain(row):
    poetry = str(row['Sloka'])
    prose = str(row['Prose'])
    
    # Split poetry and prose into words
    poetry_words = poetry.split()
    prose_words = prose.split()

    print(poetry_words)
    print(prose_words)

    # List to store matched substrings
    substrings = []

    # Perform fuzzy matching
    for poetry_word in poetry_words:
        res = []
        word_pos = {}
        for prose_word in prose_words:
            score = fuzz.partial_ratio(poetry_word, prose_word) 
            if score > 60:  
                pos = find_partial_match_start(prose_word, poetry_word)
                word_pos[prose_word] = pos
    
        sorted_items = sorted(
        ((word, position) for word, position in word_pos.items() if position is not None),
        key=lambda item: item[1])
        res = [word for word, position in sorted_items]
        substrings.extend(res)
    
    
    return " ".join(substrings) 

# Apply the function to each row
df['Segmented-Sloka'] = df.apply(match_and_retain, axis=1)

# Save the results to a new CSV
df.to_csv(output_file, index=False)

print(f"Matched substrings written to '{output_file}'")
```

### Comparing #tokens in poetry and prose

```python
import pandas as pd

# Load the test dataset
test_data = pd.read_csv('train-seg-sloka.csv')

# Function to count the number of tokens in a string
def count_tokens(text):
    return len(str(text).split())

# Compare tokens in columns 1 and 2 and identify rows where col1 < col2
condition_rows = []
for index, row in test_data.iterrows():
    col1_tokens = count_tokens(row.iloc[0])  # Assuming col1 is the first column
    col2_tokens = count_tokens(row.iloc[1])  # Assuming col2 is the second column
    if col1_tokens < col2_tokens:
        condition_rows.append((index + 2, row.iloc[0], row.iloc[1]))  # Store row number (1-based) and data

# Convert the results into a DataFrame for display
result_df = pd.DataFrame(condition_rows, columns=["Row Number", "Col1", "Col2"])

# Total number of rows matching the condition
total_condition_rows = len(condition_rows)

# Save the DataFrame to a CSV file
result_df.to_csv('filtered_rows.csv', index=False)
```

```python
import evaluate

bleu = evaluate.load("bleu")

file_path = 'test_data.csv'
data = pd.read_csv(file_path)

# Ensure the required columns are present
if "generated_Prose" not in data.columns or "ground_Prose" not in data.columns:
    raise ValueError("The CSV file must contain 'generated_Prose' and 'ground_prose' columns.")


bleu_scores = []

# Compute BLEU scores for each row
for _, row in data.iterrows():
    # Get the ground truth and generated prose
    ground_truth = row["ground_Prose"]
    generated = row["generated_Prose"]
    
    # Format as required by the BLEU computation
    references = [[ground_truth]]  # Wrap ground truth in a list
    predictions = [generated]      # Wrap generated text in a list
    
    # Compute BLEU score
    result = bleu.compute(predictions=predictions, references=references)
    bleu_scores.append(result["bleu"])

# Add the BLEU scores to the dataframe
data["BLEU_Score"] = bleu_scores

# Compute the average BLEU score
average_bleu = sum(bleu_scores) / len(bleu_scores)

# Print the average BLEU score
print(f"Average BLEU Score: {average_bleu}")

# Save the results to a new CSV file
output_path = "test_output_anvaya_bleu.csv"  # Replace with your desired output file path
data.to_csv(output_path, index=False)
print(f"Results saved to {output_path}")
```
